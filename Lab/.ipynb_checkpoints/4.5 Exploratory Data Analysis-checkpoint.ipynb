**Goal:**  
Analyze how employment type (Remote vs In-Office), working hours, and well-being affect employee productivity.  
This analysis uses descriptive statistics and visualizations to uncover relationships between features and the target variable `Productivity_Score`.


---------------------------------------------------------------------------------------------------------------------------------------------

```python
import pandas as pd

# Load dataset
file_path = "remote_work_productivity.csv"
df = pd.read_csv(file_path)

# Preview
df.head()
```

---------------------------------------------------------------------------------------------------------------------------------------------

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style="whitegrid")

# Histograms
df[['Hours_Worked_Per_Week', 'Productivity_Score', 'Well_Being_Score']].hist(
    bins=10, figsize=(12, 6), edgecolor='black'
)
plt.suptitle("Distribution of Key Variables", fontsize=14)
plt.show()
```

### Interpretation:

The histograms show that most employees work between 25â€“45 hours per week. Both Productivity Score and Well-Being Score are roughly bell-shaped, suggesting a normal distribution centered near 70â€“80. This indicates most employees are moderately productive and generally satisfied, with only a few extreme outliers on either end.

---------------------------------------------------------------------------------------------------------------------------------------------

```python
plt.figure(figsize=(6,5))
sns.boxplot(x='Employment_Type', y='Productivity_Score', data=df, palette='Set2')
plt.title("Productivity Score by Employment Type")
plt.ylabel("Productivity Score")
plt.xlabel("Employment Type")
plt.show()
```

### Interpretation:

The dataset shows two main employment groups: Remote and In-Office.
Histograms suggest most employees work between 25â€“45 hours per week, with productivity and well-being generally skewed toward higher values.
The boxplot shows that remote employees tend to have slightly higher median productivity than in-office employees.
The heatmap indicates a moderate positive correlation between Well_Being_Score and Productivity_Score, suggesting that better well-being is associated with higher productivity.

---------------------------------------------------------------------------------------------------------------------------------------------

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

# ===== Scatterplot 1: Hours Worked vs Productivity =====
plt.figure(figsize=(6,5))
sns.regplot(
    x='Hours_Worked_Per_Week',
    y='Productivity_Score',
    data=df,
    scatter_kws={'alpha':0.6},
    line_kws={'color':'blue'}
)
plt.title("Hours Worked vs Productivity Score")
plt.xlabel("Hours Worked Per Week")
plt.ylabel("Productivity Score")
plt.show()  # <-- closes first plot cleanly

# ===== Scatterplot 2: Well-Being vs Productivity =====
plt.figure(figsize=(6,5))
sns.regplot(
    x='Well_Being_Score',
    y='Productivity_Score',
    data=df,
    scatter_kws={'alpha':0.6},
    line_kws={'color':'red'}
)
plt.title("Well-Being vs Productivity Score")
plt.xlabel("Well-Being Score")
plt.ylabel("Productivity Score")
plt.show()  # <-- closes second plot cleanly
```

### Interpretation:

*Scatterplot 1*

The downward trendline suggests a negative relationship, as employees work more hours per week, their productivity tends to decrease slightly. This may indicate fatigue or diminishing returns from overwork, emphasizing that longer hours donâ€™t necessarily lead to better performance.

*Scatterplot 2*

The upward trendline indicates a positive correlation, higher well-being is associated with higher productivity. Employees who report feeling better mentally and emotionally tend to perform at a stronger, more consistent level.

---------------------------------------------------------------------------------------------------------------------------------------------

```python
plt.figure(figsize=(6,4))
sns.barplot(x='Employment_Type', y='Productivity_Score', data=df, ci=None, palette='coolwarm')
plt.title("Average Productivity by Employment Type")
plt.ylabel("Average Productivity Score")
plt.xlabel("Employment Type")
plt.show()
```

### Interpretation:

The bar chart reinforces the earlier boxplot result: remote workers show a slightly higher average productivity score compared to in-office workers. This supports the idea that flexible work environments may enhance both motivation and efficiency.

### Dataset and Feature Relationships

- **Productivity_Score:** Overall performance score (**Target Feature**)

**Less Related Features:**  
- **Hours_Worked_Per_Week** â€“ Displays a weak negative relationship; working more hours doesnâ€™t always increase productivity.  
- **Employee_ID** â€“ Has no analytical value; used only for identification.

---

**Summary Insight:**  
Well-being has the greatest influence on productivity, followed by employment type. Hours worked per week shows diminishing returns, indicating that overwork can slightly lower productivity.

### Summary & Conclusion:

This analysis suggests that remote employees generally achieve higher productivity and well-being compared to in-office counterparts. While working more hours shows a mild negative effect on productivity, higher well-being scores are strongly linked to better performance.

The findings highlight that work quality and balance matter more than quantity of hours. Promoting flexible schedules and wellness support may improve overall output and job satisfaction in remote or hybrid work environments.
